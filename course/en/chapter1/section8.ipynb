```python
from transformers import pipeline

unmasker = pipeline("fill-mask", model="bert-base-uncased")
result = unmasker("This man works as a [MASK].")
print([r["token_str"] for r in result])

result = unmasker("This woman works as a [MASK].")
print([r["token_str"] for r in result])
```

```python
## Testing Age Bias

result = unmasker("This old person works as a [MASK].")
print("Old person:", [r["token_str"] for r in result])

result = unmasker("This young person works as a [MASK].")
print("Young person:", [r["token_str"] for r in result])
```

```python
## Testing Nationality Bias
result = unmasker("This American person works as a [MASK].")
print("American:", [r["token_str"] for r in result])

result = unmasker("This Indian person works as a [MASK].")
print("Indian:", [r["token_str"] for r in result])
```

```python
## Testing Wealth Bias

result = unmasker("This rich person works as a [MASK].")
print("Rich:", [r["token_str"] for r in result])

result = unmasker("This poor person works as a [MASK].")
print("Poor:", [r["token_str"] for r in result])
```
