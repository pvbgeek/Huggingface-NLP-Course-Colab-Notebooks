```python
import os
import time
import logging

import torch
import gradio as gr
from transformers import pipeline
from transformers.utils import logging as hf_logging

os.environ["GRADIO_DEBUG"] = "1"
os.environ["TRANSFORMERS_VERBOSITY"] = "info"

logging.basicConfig(level=logging.INFO)
hf_logging.set_verbosity_info()
logging.getLogger("gradio").setLevel(logging.INFO)

print("=" * 80)
print("Libraries imported successfully.")
print("Gradio version:", gr.__version__)
print("Torch version:", torch.__version__)
print("CUDA available:", torch.cuda.is_available())

device = 0 if torch.cuda.is_available() else -1

if torch.cuda.is_available():
    print("GPU name:", torch.cuda.get_device_name(0))
else:
    print("No GPU found. Running on CPU.")

print("Using device:", "GPU" if device == 0 else "CPU")
print("=" * 80)
```

```python
def greet(name):
    print("=" * 80)
    print("Greeting function called.")
    print("Received name:", name)

    if name is None or name.strip() == "":
        print("Empty name received.")
        print("=" * 80)
        return "Please enter your name."

    response = "Hello " + name

    print("Response:", response)
    print("=" * 80)

    return response


demo = gr.Interface(
    fn=greet,
    inputs=gr.Textbox(
        label="Name",
        placeholder="Type your name here"
    ),
    outputs=gr.Textbox(
        label="Greeting"
    ),
    title="Basic Greeting Demo",
    description="Enter your name and the app will greet you."
)

print("Launching basic greeting demo...")
demo.launch(
    share=True,
    debug=True,
    show_error=True
)
```

```python
def greet_custom(name):
    print("=" * 80)
    print("Customized greeting function called.")
    print("Received name:", name)

    if name is None or name.strip() == "":
        print("Empty name received.")
        print("=" * 80)
        return "Please enter your name."

    response = "Hello " + name

    print("Response:", response)
    print("=" * 80)

    return response


textbox = gr.Textbox(
    label="Type your name here:",
    placeholder="John Doe",
    lines=2
)

custom_demo = gr.Interface(
    fn=greet_custom,
    inputs=textbox,
    outputs=gr.Textbox(label="Greeting"),
    title="Customized Textbox Greeting Demo",
    description="This demo uses a customized Gradio Textbox component."
)

print("Launching customized greeting demo...")
custom_demo.launch(
    share=True,
    debug=True,
    show_error=True
)
```

```python
print("=" * 80)
print("Loading text generation model...")
print("Model: distilgpt2")
print("=" * 80)

text_generator = pipeline(
    task="text-generation",
    model="distilgpt2",
    device=device
)

print("=" * 80)
print("Text generation model loaded successfully.")
print("Using device:", "GPU" if device == 0 else "CPU")
print("=" * 80)
```

```python
def predict(prompt):
    print("=" * 80)
    print("Text generation function called.")
    print("Received prompt:", prompt)

    if prompt is None or prompt.strip() == "":
        print("Empty prompt received.")
        print("=" * 80)
        return "Please enter a prompt."

    print("Generating completion...")
    start_time = time.time()

    output = text_generator(
        prompt,
        max_new_tokens=50,
        do_sample=True,
        temperature=0.8,
        top_p=0.95,
        pad_token_id=text_generator.tokenizer.eos_token_id
    )

    completion = output[0]["generated_text"]

    end_time = time.time()

    print("Generated completion:")
    print(completion)
    print("Processing time:", round(end_time - start_time, 2), "seconds")
    print("=" * 80)

    return completion


text_demo = gr.Interface(
    fn=predict,
    inputs=gr.Textbox(
        lines=4,
        label="Prompt",
        placeholder="Type something like: The future of AI is"
    ),
    outputs=gr.Textbox(
        lines=8,
        label="Generated Text"
    ),
    title="Text Generation Demo",
    description="Enter a prompt and DistilGPT2 will continue it.",
    examples=[
        ["The future of AI is"],
        ["Once upon a time, there was a robot who"],
        ["Machine learning helps people by"]
    ]
)

print("Launching text generation demo...")
text_demo.launch(
    share=True,
    debug=True,
    show_error=True
)
```
