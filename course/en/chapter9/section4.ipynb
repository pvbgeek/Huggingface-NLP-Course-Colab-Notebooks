```python
import os
import time
import logging
from pathlib import Path

import torch
import gradio as gr
import numpy as np
from PIL import Image

from transformers import pipeline
from transformers.utils import logging as hf_logging

os.environ["GRADIO_DEBUG"] = "1"
os.environ["TRANSFORMERS_VERBOSITY"] = "info"

logging.basicConfig(level=logging.INFO)
hf_logging.set_verbosity_info()
logging.getLogger("gradio").setLevel(logging.INFO)

print("=" * 80)
print("Libraries imported successfully.")
print("Gradio version:", gr.__version__)
print("Torch version:", torch.__version__)
print("CUDA available:", torch.cuda.is_available())

device = 0 if torch.cuda.is_available() else -1

if torch.cuda.is_available():
    print("GPU name:", torch.cuda.get_device_name(0))
else:
    print("No GPU found. Running on CPU.")

print("Using device:", "GPU" if device == 0 else "CPU")
print("=" * 80)
```

```python
print("=" * 80)
print("Loading small text-generation model for Rick bot...")
print("=" * 80)

rick_generator = pipeline(
    task="text-generation",
    model="distilgpt2",
    device=device
)

print("=" * 80)
print("Rick bot model loaded successfully.")
print("=" * 80)
```

```python
title = "Ask Rick a Question"

description = """
The bot answers questions in a playful Rick-style tone.
<img src="https://huggingface.co/spaces/course-demos/Rick_and_Morty_QA/resolve/main/rick.png" width=200px>
"""

article = "Inspired by the original Rick and Morty demo on Hugging Face Spaces."


def predict_rick(question):
    print("=" * 80)
    print("Received Rick bot question:")
    print(question)

    if question is None or question.strip() == "":
        print("Empty question received.")
        print("=" * 80)
        return "Ask me something, Morty."

    prompt = f"Question: {question}\nRick-style answer:"

    print("Generating Rick-style answer...")
    start_time = time.time()

    output = rick_generator(
        prompt,
        max_new_tokens=50,
        do_sample=True,
        temperature=0.9,
        top_p=0.95,
        pad_token_id=rick_generator.tokenizer.eos_token_id
    )

    answer = output[0]["generated_text"]

    end_time = time.time()

    print("Generated answer:")
    print(answer)
    print("Processing time:", round(end_time - start_time, 2), "seconds")
    print("=" * 80)

    return answer


rick_demo = gr.Interface(
    fn=predict_rick,
    inputs=gr.Textbox(
        lines=2,
        label="Ask Rick",
        placeholder="Example: What are you doing?"
    ),
    outputs=gr.Textbox(
        lines=6,
        label="Rick Bot Answer"
    ),
    title=title,
    description=description,
    article=article,
    examples=[
        ["What are you doing?"],
        ["Where should we time travel to?"],
        ["Why is science important?"]
    ]
)

print("Launching Rick bot demo...")
rick_demo.launch(
    share=True,
    debug=True,
    show_error=True
)
```

```python
print("=" * 80)
print("Loading image classification model...")
print("=" * 80)

image_classifier = pipeline(
    task="image-classification",
    model="google/vit-base-patch16-224",
    device=device
)

print("=" * 80)
print("Image classification model loaded successfully.")
print("=" * 80)
```

```python
def classify_image(input_image):
    print("=" * 80)
    print("Received image for classification.")

    if input_image is None:
        print("No image received.")
        print("=" * 80)
        return {}

    print("Input image type:", type(input_image))
    print("Input image size:", input_image.size)

    print("Classifying image...")
    start_time = time.time()

    predictions = image_classifier(input_image)

    end_time = time.time()

    result = {
        item["label"]: float(item["score"])
        for item in predictions[:5]
    }

    print("Top predictions:")
    print(result)
    print("Processing time:", round(end_time - start_time, 2), "seconds")
    print("=" * 80)

    return result


image_demo = gr.Interface(
    fn=classify_image,
    inputs=gr.Image(
        type="pil",
        label="Upload Image"
    ),
    outputs=gr.Label(
        num_top_classes=5,
        label="Prediction"
    ),
    title="Image Classification Demo",
    description="Upload an image and the model will predict the object/class."
)

print("Launching image classification demo...")
image_demo.launch(
    share=True,
    debug=True,
    show_error=True
)
```

```python
from torch import nn

print("=" * 80)
print("Loading sketch recognition model...")
print("=" * 80)

LABELS = Path("class_names.txt").read_text().splitlines()

print("Number of labels:", len(LABELS))
print("First few labels:", LABELS[:10])

sketch_model = nn.Sequential(
    nn.Conv2d(1, 32, 3, padding="same"),
    nn.ReLU(),
    nn.MaxPool2d(2),

    nn.Conv2d(32, 64, 3, padding="same"),
    nn.ReLU(),
    nn.MaxPool2d(2),

    nn.Conv2d(64, 128, 3, padding="same"),
    nn.ReLU(),
    nn.MaxPool2d(2),

    nn.Flatten(),
    nn.Linear(1152, 256),
    nn.ReLU(),
    nn.Linear(256, len(LABELS)),
)

state_dict = torch.load(
    "pytorch_model.bin",
    map_location="cuda" if torch.cuda.is_available() else "cpu"
)

sketch_model.load_state_dict(state_dict, strict=False)
sketch_model.eval()

if torch.cuda.is_available():
    sketch_model = sketch_model.cuda()
    print("Sketch model moved to GPU.")
else:
    print("Sketch model running on CPU.")

print("Sketch recognition model loaded successfully.")
print("=" * 80)
```

```python
def extract_image_from_editor(editor_value):
    if editor_value is None:
        return None

    if isinstance(editor_value, dict):
        if editor_value.get("composite") is not None:
            return editor_value["composite"]
        if editor_value.get("background") is not None:
            return editor_value["background"]

    return editor_value


def predict_sketch(editor_value):
    print("=" * 80)
    print("Received sketch input.")

    image = extract_image_from_editor(editor_value)

    if image is None:
        print("No sketch received.")
        print("=" * 80)
        return {}

    if isinstance(image, np.ndarray):
        image = Image.fromarray(image)

    print("Original sketch image size:", image.size)

    image = image.convert("L")
    image = image.resize((28, 28))

    x = np.array(image).astype("float32")
    x = 255.0 - x
    x = x / 255.0

    tensor = torch.tensor(x, dtype=torch.float32).unsqueeze(0).unsqueeze(0)

    if torch.cuda.is_available():
        tensor = tensor.cuda()

    print("Tensor shape:", tensor.shape)
    print("Predicting sketch label...")

    start_time = time.time()

    with torch.no_grad():
        out = sketch_model(tensor)

    probabilities = torch.nn.functional.softmax(out[0], dim=0)
    values, indices = torch.topk(probabilities, 5)

    end_time = time.time()

    result = {
        LABELS[index.item()]: value.item()
        for index, value in zip(indices, values)
    }

    print("Top predictions:")
    print(result)
    print("Processing time:", round(end_time - start_time, 2), "seconds")
    print("=" * 80)

    return result


sketch_demo = gr.Interface(
    fn=predict_sketch,
    inputs=gr.ImageEditor(
        type="pil",
        label="Draw Here"
    ),
    outputs=gr.Label(
        num_top_classes=5,
        label="Prediction"
    ),
    title="Sketch Recognition",
    description="Draw a common object like a shovel, laptop, or cup. The model will guess the sketch.",
    article="<p style='text-align: center'>Sketch Recognition | Demo Model</p>",
    live=False
)

print("Launching sketch recognition demo...")
sketch_demo.launch(
    share=True,
    debug=True,
    show_error=True
)
```
