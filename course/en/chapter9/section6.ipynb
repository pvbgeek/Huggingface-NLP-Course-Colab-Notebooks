```python
import random
import gradio as gr


def chat(message, history):
    if message.startswith("How many"):
        return str(random.randint(1, 10))

    elif message.startswith("How"):
        return random.choice(["Great", "Good", "Okay", "Bad"])

    elif message.startswith("Where"):
        return random.choice(["Here", "There", "Somewhere"])

    else:
        return "I don't know"


demo = gr.ChatInterface(
    fn=chat,
    title="Simple Rule-Based Chatbot"
)

demo.launch()
```

```python
import requests
import tensorflow as tf
import gradio as gr


# Load pretrained MobileNetV2 model
inception_net = tf.keras.applications.MobileNetV2()


# Download human-readable ImageNet labels
response = requests.get("https://git.io/JJkYN")
labels = response.text.split("\n")


def classify_image(inp):
    # Resize safely in case uploaded image is not exactly 224x224
    inp = tf.image.resize(inp, (224, 224)).numpy()

    # Add batch dimension
    inp = inp.reshape((-1, 224, 224, 3))

    # Preprocess for MobileNetV2
    inp = tf.keras.applications.mobilenet_v2.preprocess_input(inp)

    # Predict probabilities
    prediction = inception_net.predict(inp).flatten()

    # Return label-probability dictionary
    return {labels[i]: float(prediction[i]) for i in range(1000)}


image = gr.Image(
    width=224,
    height=224,
    type="numpy",
    label="Upload Image"
)

label = gr.Label(
    num_top_classes=3,
    label="Prediction"
)

title = "Gradio Image Classification Example"

demo = gr.Interface(
    fn=classify_image,
    inputs=image,
    outputs=label,
    title=title
)

demo.launch()
```
