```python
import os
import time
import logging
import threading

os.environ["GRADIO_DEBUG"] = "1"
os.environ["TRANSFORMERS_VERBOSITY"] = "info"

import torch
import gradio as gr
from transformers import pipeline
from transformers.utils import logging as hf_logging

logging.basicConfig(level=logging.INFO)

hf_logging.set_verbosity_info()
logging.getLogger("gradio").setLevel(logging.INFO)

print("=" * 80)
print("Libraries imported successfully.")
print("Gradio version:", gr.__version__)
print("Torch version:", torch.__version__)
print("CUDA available:", torch.cuda.is_available())

if torch.cuda.is_available():
    print("GPU name:", torch.cuda.get_device_name(0))
else:
    print("No GPU found. Code will run on CPU.")

print("Debug logs enabled.")
print("=" * 80)
```

```python
device = 0 if torch.cuda.is_available() else -1

print("=" * 80)
print("Loading text generation model: distilgpt2")
print("Using device:", "GPU" if device == 0 else "CPU")
print("=" * 80)

generator = pipeline(
    task="text-generation",
    model="distilgpt2",
    device=device
)

print("=" * 80)
print("Model loaded successfully.")
print("=" * 80)
```

```python
text_done_event = threading.Event()
text_demo = None


def generate_text(input_text):
    global text_demo

    print("=" * 80)
    print("Received input text:")
    print(input_text)

    if input_text is None or input_text.strip() == "":
        print("Empty input received.")
        print("=" * 80)
        return "Please enter some starting text."

    print("Generating output...")

    output = generator(
        input_text,
        max_new_tokens=50,
        do_sample=True,
        temperature=0.8,
        top_p=0.95,
        pad_token_id=generator.tokenizer.eos_token_id
    )

    generated_text = output[0]["generated_text"]

    print("Generated output:")
    print(generated_text)
    print("=" * 80)

    print("Marking text generation app as complete.")
    text_done_event.set()

    return generated_text


examples = [
    ["Artificial intelligence will change education because"],
    ["Once upon a time, there was a robot who"],
    ["The future of machine learning is"],
    ["The Moon's orbit around Earth has"],
]

text_demo = gr.Interface(
    fn=generate_text,
    inputs=gr.Textbox(
        lines=5,
        label="Input Text",
        placeholder="Type something like: Artificial intelligence will change education because"
    ),
    outputs=gr.Textbox(
        lines=8,
        label="Generated Text"
    ),
    title="Small Text Generation Demo",
    description="Type starting text and the model will continue it. After one successful generation, this app will close automatically.",
    examples=examples
)

print("=" * 80)
print("Launching text generation app...")
print("Submit one input. After generation finishes, the app will close automatically.")
print("=" * 80)

text_demo.launch(
    debug=True,
    show_error=True,
    prevent_thread_lock=True
)

print("Waiting for one text generation request to finish...")

while not text_done_event.is_set():
    time.sleep(1)

print("Text generation completed.")
print("Closing text generation app...")

text_demo.close()

print("Text generation app closed. Cell completed.")
```

```python
import gradio as gr
from rembg import remove
from PIL import Image
import io
import time

print("=" * 80)
print("Local background removal app starting...")
print("Using rembg locally instead of Hugging Face Space.")
print("=" * 80)


def remove_background_upload(input_image):
    print("=" * 80)
    print("Received uploaded image.")

    if input_image is None:
        print("No image received.")
        print("=" * 80)
        return None

    print("Input image type:", type(input_image))
    print("Input image size:", input_image.size)

    print("Removing background...")
    start_time = time.time()

    output_image = remove(input_image)

    end_time = time.time()

    print("Background removed successfully.")
    print("Output image type:", type(output_image))
    print("Output image size:", output_image.size)
    print("Processing time:", round(end_time - start_time, 2), "seconds")
    print("=" * 80)

    return output_image


bg_demo = gr.Interface(
    fn=remove_background_upload,
    inputs=gr.Image(
        type="pil",
        label="Upload Image"
    ),
    outputs=gr.Image(
        type="pil",
        label="Image Without Background"
    ),
    title="Local Background Removal",
    description="Upload an image. The app will remove the background locally using rembg."
)

print("Launching local background removal app...")
bg_demo.launch(debug=True, show_error=True)
```

```python
import gradio as gr
from rembg import remove
import time

print("=" * 80)
print("Local webcam background removal app starting...")
print("Trying inline Colab display with share=True.")
print("=" * 80)


def remove_background_webcam(input_image):
    print("=" * 80)
    print("Received webcam image.")

    if input_image is None:
        print("No webcam image received.")
        print("=" * 80)
        return None

    print("Input image type:", type(input_image))
    print("Input image size:", input_image.size)

    print("Removing webcam background...")
    start_time = time.time()

    output_image = remove(input_image)

    end_time = time.time()

    print("Webcam background removed successfully.")
    print("Output image type:", type(output_image))
    print("Output image size:", output_image.size)
    print("Processing time:", round(end_time - start_time, 2), "seconds")
    print("=" * 80)

    return output_image


webcam_demo = gr.Interface(
    fn=remove_background_webcam,
    inputs=gr.Image(
        sources=["webcam"],
        type="pil",
        label="Webcam Input"
    ),
    outputs=gr.Image(
        type="pil",
        label="Image Without Background"
    ),
    title="Local Webcam Background Removal",
    description="If webcam permission does not appear inline, open the public gradio.live link in a new tab."
)

print("Launching local webcam background removal app...")

webcam_demo.launch(
    debug=True,
    show_error=True,
    share=True,
    inline=True,
    inbrowser=False
)
```
