```python
# Explore what's inside BertConfig
from transformers import BertConfig

config = BertConfig()
print("Vocabulary size      :", config.vocab_size)
print("Hidden size          :", config.hidden_size)
print("Number of layers     :", config.num_hidden_layers)
print("Attention heads      :", config.num_attention_heads)
print("Intermediate size    :", config.intermediate_size)
print("Max position embeddings:", config.max_position_embeddings)
```

```python
# Create a smaller custom BERT — useful for low resource environments
from transformers import BertConfig, BertModel

small_config = BertConfig(
    vocab_size=30522,
    hidden_size=256,          # Default is 768 — much smaller!
    num_hidden_layers=4,      # Default is 12
    num_attention_heads=4,    # Default is 12
    intermediate_size=512,    # Default is 3072
)

small_model = BertModel(small_config)
print("Small model config:")
print(f"  Hidden size    : {small_config.hidden_size}")
print(f"  Layers         : {small_config.num_hidden_layers}")
print(f"  Attention heads: {small_config.num_attention_heads}")

# Count parameters
total_params = sum(p.numel() for p in small_model.parameters())
print(f"\nTotal parameters: {total_params:,}")

# Compare with standard BERT
standard_model = BertModel(BertConfig())
standard_params = sum(p.numel() for p in standard_model.parameters())
print(f"Standard BERT parameters: {standard_params:,}")
```

```python
# Understand the two main outputs of BERT
import torch
from transformers import BertModel

model = BertModel.from_pretrained("bert-base-uncased")

sample_input = torch.tensor([[101, 7592, 999, 102]])  # "hello !"
output = model(sample_input)

print("last_hidden_state shape:", output.last_hidden_state.shape)
# [batch_size, sequence_length, hidden_size]
# → one vector per TOKEN

print("pooler_output shape    :", output.pooler_output.shape)
# [batch_size, hidden_size]
# → one vector for the entire SENTENCE (from [CLS] token)
```

```python
# Save your custom config and reload it
import os
from transformers import BertConfig, BertModel

# Create and save
small_config = BertConfig(hidden_size=256, num_hidden_layers=4, num_attention_heads=4)
small_model = BertModel(small_config)
small_model.save_pretrained("my_small_bert")

# Reload
reloaded_model = BertModel.from_pretrained("my_small_bert")
reloaded_config = reloaded_model.config

print("Reloaded hidden size    :", reloaded_config.hidden_size)
print("Reloaded layers         :", reloaded_config.num_hidden_layers)
print("Reloaded attention heads:", reloaded_config.num_attention_heads)
print("\nFiles saved:", os.listdir("my_small_bert"))
```

```python
# Run a batch of sequences and inspect outputs
import torch
from transformers import BertModel

model = BertModel.from_pretrained("bert-base-uncased")

# 3 sequences of length 4
encoded_sequences = [
    [101, 7592, 999, 102],
    [101, 4658, 1012, 102],
    [101, 3835, 999, 102],
]

batch_input = torch.tensor(encoded_sequences)
output = model(batch_input)

print("Batch size      :", output.last_hidden_state.shape[0])
print("Sequence length :", output.last_hidden_state.shape[1])
print("Hidden size     :", output.last_hidden_state.shape[2])
print("\nFull output shape:", output.last_hidden_state.shape)
# Each of the 3 sentences gets a 4x768 matrix of contextual embeddings
```
